## mtcars.csv dataset

```python
#importing the required library - pandas
import pandas as pd
```

```python
#loading the data from the csv file as a dataframe 
x = pd.read_csv('mtcars.csv')

#displaying the loaded dataset in the form of a dataframe
x
```

### Q1)  Tabulate number of cylinders in the dataset.

```python
#tabulate the number of cylinders in the dataset
cylinders = x['cyl'].value_counts()

#print the result
print("total number of cylinders of each type:")
print(cylinders)
```

```python
#calculate the total number of cylinders
total_cylinders = sum(cylinders)

#print the result
print("Total number of cylinders:", total_cylinders)
```

### Q2)  Find the five number summary of milespergallon.

### Q3) Draw a histogram for hp.

```python
#importing the required library - pyplot(matplotlib)
import matplotlib.pyplot as plt
```

```python
#draw a histogram of the hp variable
plt.hist(x['hp'], bins=10, edgecolor='black')

#adding labels and title to the plot
plt.xlabel('Horsepower')
plt.ylabel('Frequency')
plt.title('Histogram of Horsepower')

#show the plot
plt.show()
```

### Q6) Find the car with the minimum displacement.

### Q12) Do side by side boxplot of mpg for cars with am (automatic transmission) Vs cars with Manual transmission.

```python
#for automatic transmission, am = 0 and for manual transmission, am = 1

#creating a list of dataframes for cars with am = 0 and am = 1
data = [x[x['am'] == 0]['mpg'], x[x['am'] == 1]['mpg']]

#create a boxplot for the data
fig = plt.figure(figsize=(10, 5))
bp = plt.boxplot(data, patch_artist=True, notch=False)

#add colors to the boxplot
colors = ['pink', 'lightblue']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

#add labels and title
plt.xticks([1, 2], ['Manual', 'Automatic'])
plt.ylabel('Miles per gallon')
plt.title('MPG for cars with manual VS automatic transmission')

#show the plot
plt.show()
```
