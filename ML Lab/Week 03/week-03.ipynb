## Q.1) Draw the graphs asked in week_2 mtcars & cereals dataset

```python
#importing the required library - pandas
import pandas as pd

#loading the data from the csv file as a dataframe 
x = pd.read_csv('mtcars.csv')

#displaying the loaded dataset in the form of a dataframe
x
```

### Draw a histogram for hp

```python
#importing the required library - pyplot(matplotlib)
import matplotlib.pyplot as plt

#draw a histogram of the hp variable
plt.hist(x['hp'], bins=10, edgecolor='black')

#adding labels and title to the plot
plt.xlabel('Horsepower')
plt.ylabel('Frequency')
plt.title('Histogram of Horsepower')

#show the plot
plt.show()
```

### Box plot the miles per gallon

```python
#draw a box plot of the milespergall variable
plt.boxplot(x['mpg'])

#add labels and title to the plot
plt.xlabel('Miles per gallon')
plt.title('Box Plot of Miles per gallon')

#show the plot
plt.show()
```

### Do side by side boxplot of mpg for cars with am (automatic transmission) Vs cars with Manual transmission.

```python
#for automatic transmission, am = 0 and for manual transmission, am = 1

#creating a list of dataframes for cars with am = 0 and am = 1
data = [x[x['am'] == 0]['mpg'], x[x['am'] == 1]['mpg']]

#create a boxplot for the data
fig = plt.figure(figsize=(10, 5))
bp = plt.boxplot(data, patch_artist=True, notch=False)

#add colors to the boxplot
colors = ['pink', 'lightblue']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

#add labels and title
plt.xticks([1, 2], ['Manual', 'Automatic'])
plt.ylabel('Miles per gallon')
plt.title('MPG for cars with manual VS automatic transmission')

#show the plot
plt.show()
```

## Q.2) Generate a dataset of 20+ random number. Apply suitable dataset to draw the following plots:

### a) Histograms, Line plots in a 2-D plane. b) Bar charts: Normal, Side by side bar charts, Stacked Bar charts. c) Sine curves using math.pi, range and arange functions. d) 3D Scatter and Surface plots using matplotlib and seaborn packages. e) Violin plots using Seaborn. f) Pie charts. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Generate a dataset of 20 random numbers
np.random.seed(123)
data = pd.DataFrame({'A': np.random.normal(0, 1, 20),
                     'B': np.random.normal(1, 1, 20),
                     'C': np.random.normal(2, 1, 20)})
data
```

```python
# Histograms and Line plots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
data.plot(kind='hist', bins=10, alpha=0.5, ax=axs[0])
data.plot(kind='line', ax=axs[1])
plt.show()
```

```python
# Bar charts
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
data.mean().plot(kind='bar', ax=axs[0], title='Normal Bar Chart')
data.mean().plot(kind='bar', ax=axs[1], title='Side-by-Side Bar Chart', position=0.5, width=0.3)
data.plot(kind='bar', stacked=True, ax=axs[2], title='Stacked Bar Chart')
plt.show()
```

```python
#sine curve
x = np.arange(0, 2 * math.pi, 0.1)
y = np.sin(x)

#Plot the curve
plt.plot(x, y)
plt.show()
```

```python
#create 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['A'], data['B'], data['C'], c='r', marker='o')
plt.show()

#create 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x, y = np.meshgrid(data['A'], data['B'])
z = np.sin(np.sqrt(x**2 + y**2))
ax.plot_surface(x, y, z, cmap='viridis')
plt.show()
```

```python
#Violin plots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
sns.violinplot(data=data, ax=axs[0])
sns.violinplot(data=data, ax=axs[1], inner='stick')
plt.show()
```

```python
#pie charts
data.mean().plot(kind='pie', figsize=(6, 6), autopct='%1.1f%%')
plt.show()
```

## Q.4) Create dataset of student marks in 5 subjects program wise (CSE, ECE, EEE, ICT programs) with for atleast 10 students each. Draw all suitable charts for this data.

```python
programs = ['CSE', 'ECE', 'EEE', 'ICT']
subjects = ['Maths', 'DSA', 'OOPs', 'Data Analytics', 'Machine Learning']
num_students = 10
min_marks = 40
max_marks = 100

df = pd.DataFrame()
for program in programs:
    marks = np.random.randint(min_marks, max_marks+1, size=(num_students, len(subjects)))
    df_program = pd.DataFrame(marks, columns=subjects)
    df_program['Program'] = program
    df = df.append(df_program, ignore_index=True)

df_grouped = df.groupby('Program')
df_mean = df_grouped.mean()
df_std = df_grouped.std()
```

```python
df_mean.plot(kind='bar', yerr=df_std, capsize=4)
plt.title('Mean Marks by Program and Subject')
plt.xlabel('Program')
plt.ylabel('Marks')
plt.show()
```

```python
df.boxplot(by='Program', column=subjects)
plt.title('Marks by Program and Subject')
plt.suptitle('')
plt.xlabel('Program')
plt.ylabel('Marks')
plt.show()
```

```python
sns.violinplot(x='Program', y='value', hue='variable', data=pd.melt(df, id_vars=['Program'], value_vars=subjects),
               split=False, inner='stick', palette='muted')
sns.despine(left=True)
plt.title('Marks by Program and Subject')
plt.xlabel('Program')
plt.ylabel('Marks')
plt.show()
```

## Q.5) Create a matrix using numpy. Perform the matrix transpose, inverse and multiplication. 

```python
m1 = np.array([[4, 2, 8], [3, 6, 2], [8, 4, 3]])

print("Original Matrix: \n", m1)

t = np.transpose(m1)

print("\nTranspose of Matrix: \n", t)

i = np.linalg.inv(m1)

print("\nInverse of Matrix: \n", i)

m2 = np.array([[3, 5, 3], [1, 8, 3], [4, 8, 7]])

mul = np.matmul(m1, m2)

print("\nMultiplication of two matrices: \n", mul)
```
