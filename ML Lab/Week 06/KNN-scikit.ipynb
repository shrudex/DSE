```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import math
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
#reading the CSV file
x = pd.read_csv("KNN-dataset.csv")

#displaying the dataframe
x
```

```python
#check if there is any NULL value
x.isnull().sum()
```

```python
#the dataset has NO null values, so don't need to handle them
```

```python
#replacing 0 with the mean values

#all the columns in which value 0 can't be present
l_col = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

for i in l_col:
    x[i] = x[i].replace(0, np.NaN) #replace 0 will NULL
    mean = int(x[i].mean(skipna=True)) #calculate mean, excluding NULL values
    x[i] = x[i].replace(np.NaN, mean) #replace NULL with mean
    
#displaying the updated dataframe    
x
```

```python
#splitting training and testing dataset

#X will have all rows, but from 0 to 7th index columns
X = x.iloc[:, 0:8]

#y will have all rows, but the 8th index column
y = x.iloc[:, 8]

#splitting the dataset, with test-size = 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size = 0.3)
```

```python
#standardizing data
sd = StandardScaler()

X_train = sd.fit_transform(X_train)
X_test = sd.transform(X_test)
```

```python
#training the model 
k = math.sqrt(len(y_test))
if (math.floor(k)%2==0):
    k = math.floor(k)
else: 
    k = math.ceil(k)

classifier = KNeighborsClassifier(n_neighbors = k, p=2, metric = "euclidean")
```

```python
#fitting the model
classifier.fit(X_train, y_train)
```

```python
#predicting the test set results
y_pred = classifier.predict(X_test)
y_pred
```

```python
#creating the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#measuring the accuracy of the model
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("F Score:", f1_score(y_test, y_pred))
```
