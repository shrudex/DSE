# Titanic Dataset

##### www.kaggle.com/c/titanic/data

```python
#importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import *
from sklearn.preprocessing import StandardScaler
```

```python
#reading the CSV file
x = pd.read_csv("titanic.csv")

#displaying the dataframe
x
```

```python
#dataset size - 891 rows x 12 columns
```

```python
#check if there is any NULL value
x.isnull().sum()
```

```python
#the dataset has null values, so we will have to fill them
x['Age'].fillna(x['Age'].median(), inplace=True)
x['Embarked'] = x['Embarked'].fillna(x['Embarked'].mode()[0])

x
```

```python
#converting categorical data to numerical
x['Sex'] = x['Sex'].map({'female': 0, 'male': 1})
x['Embarked'] = x['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

```python
#dropping the unnecessary column as it is causing problem later
x = x.drop(["Name", "Ticket", "Cabin"], axis=1)
```

```python
#splitting training and testing dataset

#X will have all rows and all columns except the "Survived" column which is 1st column (index-wise)
X = x.iloc[:, [0] + list(range(2, len(x.columns)))]

#y will have all rows of the "Survived" column
y = x["Survived"]
```

```python
#standardizing the training dataset
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
#splitting the data into train and test sets for 6 combinations
ts = 0.2
for i in range(0, 6):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=ts, random_state=i*i*i)
    
    #increasing the test-size for the next iteration
    ts = ts + 0.1 
    
    #creating a model
    model = SVC()
    
    #training the model on the training set
    model.fit(X_train, y_train)
    
    #evaluating the model
    #predicting the values of the X_test dataset
    y_pred = model.predict(X_test)
    
    #finding the accuracy and f-score of the model
    accuracy = accuracy_score(y_test, y_pred)
    fscore = f1_score(y_test, y_pred)
    
    print("Accuracy score for this model is :", accuracy)
    print("F score for this model is :", fscore)
    print()
```
